```python
import numpy as np
import matplotlib.pyplot as plt
```

## Convolution

```python
# x = list(map(int,input().split()))
# h = list(map(int,input().split()))

x = [3,4,6,1]
h = [1,2,3]

l = len(x) + len(h) - 1

y = [0] * l

for i in range(len(x)):
    for j in range(len(h)):
        y[i+j] += x[i] * h[j]

print(y) 
print(np.convolve(x,h))
```

## Cross-Correlation

```python
x = [3,4,6,1]
h = [1,2,3]

l = len(x) + len(h) - 1

rh = np.flip(h)

y = [0] * l

for i in range(len(x)):
    for j in range(len(h)):
        y[i+j] += x[i] * rh[j]

print(y)
print(np.correlate(x,h,'full'))
```

## Auto Correltaion

```python
x = [3,4,6,1]
rx = np.flip(x)
l = 2*len(x) - 1

y = [0] * l

for i in range(len(x)):
    for j in range(len(x)):
        y[i+j] += x[i] * rx[j]

print(y)
print(np.correlate(x,x,'full'))
```

## Impulse

```python
x = np.arange(-5,5)
l = len(x)
y = [0]*l

y[5] = 1

plt.stem(x,y)
```

## Ramp

```python
x = np.arange(-5,5)
y = [0] * len(x)

for i in range(len(x)):
    if(x[i]>0):
        y[i] = x[i]

plt.xticks(np.arange(-5,5,1))
plt.stem(x,y)
```

```python
x = np.arange(-5,5)
y = [0] * len(x)

for i in range(len(x)):
    if(x[i]>=0):
        y[i] = 1

plt.xticks(np.arange(-5,5,1))
plt.stem(x,y)
```

## Expoential

```python

y1 = [0] * 40
y2 = [0] * 40
a1 = 0.9
a2 = 1.1

for i in range(len(y1)):
    y1[i] = a1 ** i

for i in range(len(y2)):
    y2[i] = a2 ** i

plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.stem(y1)

plt.subplot(1,2,2)
plt.stem(y2)
```

## Discrete Signal

```python
y = [9,-4,5,6,2,11,-3,0,1,5,6]
x = np.arange(-4,7)

plt.stem(x,y)
```

## Discrete Signal Shifting

```python
y = np.array([9,-4,5,6,2,11,-3,0,1,5,6])
x = np.arange(-4,7)

plt.figure(figsize=(20,4))
# plt.xticks(np.arange(-4,7,1))

plt.subplot(1,3,1)
plt.stem(x,y)
plt.xticks(x)
plt.title("Main Signal")

# left shift
xl = x + 2
plt.subplot(1,3,2)
plt.title("Main Signal")
plt.xticks(xl)
plt.stem(xl,y)

xr = x - 2
plt.subplot(1,3,3)
plt.title("Main Signal")
plt.xticks(xr)
plt.stem(xr,y)
```

## Sine Wave

```python
f = 4
Fs = 100
L = 100

t = np.arange(1,L) / Fs

y = np.sin(2* np.pi * f * t)

plt.stem(y)
```

## Composite signal

```python
f = 4
Fs = 100
L = 100
t = np.arange(1,L) / Fs

plt.figure(figsize=(20,4))

y = np.sin(2* np.pi * f * t)
plt.subplot(1,3,1)
plt.plot(y)


f1 = 9
y1 = np.sin(2 * np.pi * f1 * t)
plt.subplot(1,3,2)
plt.plot(y1)


y2 = y + y1
plt.subplot(1,3,3)
plt.plot(y2)

```
